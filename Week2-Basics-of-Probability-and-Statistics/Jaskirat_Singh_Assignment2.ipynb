```python
import numpy as np, matplotlib.pyplot as plt
```

Solution to Question 1

```python
# Part a

# Created an array consisting the velocities
velocities = np.array([420, 445, 380, 510, 398, 475, 425, 390, 460, 440])

# using the in-built method for mean
avg_speed = np.mean(velocities)
print(f"Average Speed = {avg_speed} m/s")

# Part b

# using the in-built method for standard deviation
std_speed = np.std(velocities)
print(f"Standard deviation = {std_speed} m/s\n")

# Part c

# plotting the velocities of the particles
plt.figure(figsize = (9, 6))

plt.bar(list(range(1, len(velocities)+1)), velocities, color = 'b')
plt.axhline(avg_speed, linestyle = '--', color = 'r', linewidth = 1.25, label = f"Average Speed ({avg_speed} m/s)")
plt.title("Particle Speed Distribution with Mean", fontsize = 14, fontweight = 'bold')
plt.xlabel("Particle Number", fontsize = 12)
plt.ylabel("Speed m/s", fontsize = 12)
plt.legend()
plt.grid(True, axis = 'y', alpha = 0.3)
plt.show()
```

Solution to Question 2

```python
# Part b

mean_given = 22; std_given = 0.5
temperatures = np.random.normal(loc = mean_given, scale = std_given, size = 100)
# loc most probably means location of centre (mean value for the bell curve)
# scale means the value for spread which is taken as the value of standard deviation



# Part b

# plotting the temperature values to confirm their distribution visually
plt.figure(figsize = (9, 6))

plt.hist(temperatures, bins=20, color = 'b', edgecolor = 'gray')
#plt.axvline(mean, linestyle = '--', color = 'r', linewidth = 1.25, label = 'Provided mean')
plt.axvline(np.mean(temperatures), linestyle = '--', color = 'r', linewidth = 1.25, label = 'Mean of generated data')
plt.axvline(np.mean(temperatures)-np.std(temperatures), linestyle = '-.', color = 'b', linewidth = 1.25, alpha = 0.6, label = r'Region between $-\sigma$ and $\sigma$')
plt.axvline(np.mean(temperatures)+np.std(temperatures), linestyle = '-.', color = 'b', linewidth = 1.25, alpha = 0.6)
plt.title("Distribution of Temperature Measurements")
plt.xlabel(r"Temperature ($\degree$C)")
plt.ylabel("Frequency")
plt.grid(True, axis = 'y', alpha = 0.3)
plt.legend()
plt.show()

# Below is the difference between the provided mean and the mean of the generated values
# The mean of generated values is not very far off to the given mean
print('\n'+'='*60)
print("Provided Mean = 22")
print(f"Mean of generated data = {np.mean(temperatures):.4f}")
print(f"Difference = {np.abs(22-np.mean(temperatures)):.4f}")
print('\n'+'='*60+'\n')


# Part c

# Rather than providing 100 as the total number of values, i used length of the temperatures array to maintain the generality of the code over various data
oneSigmaVals = np.array([i for i in temperatures if (i>=mean_given-std_given and i<=mean_given+std_given)])
print(f"Number of values within one standard deviation = {len(oneSigmaVals)}")
print(f"Percentage of measurements between 21.5°C and 22.5°C = {len(oneSigmaVals)*100/len(temperatures)}%")
print('\n'+'='*60)
```

Solution to Question 3

```python
# Part a

scalePara = 2.5 # Scale Parameter
velocities = np.random.exponential(scale = scalePara, size = 5000)

# Part b

listOfAverages = [np.mean(velocities[:100]), np.mean(velocities[:1000]), np.mean(velocities[:3000]), np.mean(velocities)]
"""
in the above line, i used list comprehension to create the list of averages of different
number of velocities. This saved me 4 variables and is faster
"""

# Part c

# creating list of sample spaces to  find cumulative frequency
sampleSizes = list(range(1, len(velocities)+1))

plt.figure(figsize = (15, 6))

plt.plot(sampleSizes, [np.mean(velocities[:i]) for i in sampleSizes], color = 'b')
plt.xlim(0)
plt.axhline(np.mean(velocities), linestyle = '--', color = 'r', linewidth = 1.25, label = f"{np.mean(velocities):.2f}")
plt.title("Law of Large Numbers: Convergence of Averages", fontweight = 'bold')
plt.xlabel("Number of Samples")
plt.ylabel("Average Velocity")
plt.grid(True, linestyle = '--', color = 'grey', linewidth = 0.75, alpha = 0.6)
plt.legend()
plt.show()
```

Solution to Question 4

```python
# Part a

"""
Measurements from instruments follow Normal/Gaussian Distribution
Hence generating value as normal distribution
"""
inst1 = np.random.normal(loc = 5.80, scale = 0.15, size = 500)
inst2 = np.random.normal(loc = 5.75, scale = 0.35, size = 500)

# Part b

plt.figure(figsize = (8, 6))

plt.hist(inst1, bins = 15, color = 'b', alpha = 0.5, label = "Instrument 1")
plt.hist(inst2, bins = 15, color = 'r', alpha = 0.5, label = "Instrument 2")
plt.axvline(np.mean(inst1), color = 'b', linestyle = '--', alpha = 0.6, label = f"Mean-Instrument 1 ({np.mean(inst1):.2f})")
plt.axvline(np.mean(inst2), color = 'r', linestyle = '--', alpha = 0.6, label = f"Mean-Instrument 2 ({np.mean(inst2):.2f})")
plt.title("Comparison of two Instruments", fontweight = 'bold')
plt.xlabel("Conductivity (S/m)")
plt.ylabel("Frequency")
plt.legend()
plt.grid(True, axis = 'y', linestyle = '--', color = 'grey', linewidth = 0.75, alpha = 0.6)
plt.show()

# Part c

print('\n'+'='*60+'\n')
precision_ratio = np.std(inst2)/np.std(inst1)
print(f"Instrument 2 is {precision_ratio:.2f} times more scattered than Instrument 1")
print('\n'+'='*60)
```
